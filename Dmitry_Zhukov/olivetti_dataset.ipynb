```python
from sklearn.datasets import fetch_olivetti_faces
import numpy as np
```

```python
rng = np.random.RandomState(0)
dataset = fetch_olivetti_faces(shuffle=True, random_state=rng)
faces = dataset.images
target = dataset.target
classes = np.unique(target)
classes.sort()
```

```python
faces.shape
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
#img = mpimg.imread(faces[0])
imgplot = plt.imshow(faces[0])
plt.show()
```

```python
n_row = 2
n_col = 10
fig, ax = plt.subplots(n_row, n_col, figsize=(6,3), dpi=300)
ax = ax.flatten()
for i in range(n_row * n_col):
  ax[i].imshow(X=faces[i], cmap='gray')
  ax[i].set_xticks([])
  ax[i].set_yticks([])
  ax[i].set_title("Id:{}".format(target[i]))
```
